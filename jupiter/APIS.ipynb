# Métodos HTTP

- GET:	Recuperar datos existentes
- POST:	Agregar nuevos datos
- PUT:	Actualizar datos existentes
- PATCH:	Actualizar parcialmente los datos existentes
- DELETE:	Borrar datos

API = https://fakestoreapi.com/

## GET

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

response = requests.get(f"{BASE_URL}/products")
print(response.json())
print(response.status_code)
```

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

query_params = {
    "limit": 3
}

response = requests.get(f"{BASE_URL}/products", params=query_params)
print(response.json())
print(response.url)
```

```python
import requests
BASE_URL = 'https://fakestoreapi.com'

preciominimo = int(input("Precio minimo de los productos: "))
preciomaximo = int(input("Precio maximo de los productos: "))
response = requests.get(f"{BASE_URL}/products")

resultado = []
for producto in response.json():
  if preciominimo <= int(producto['price']) <= preciomaximo:
    resultado.append(producto['title'])
print(resultado)
```

## POST

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

new_product = {
    "title": 'test product',
    "price": 13.5,
    "description": 'lorem ipsum set',
    "category": 'electronic'
}

response = requests.post(f"{BASE_URL}/products", json=new_product)
print(response.json())
```

## PUT

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

updated_product = {
    "title": 'updated_product',
    "category": 'clothing'
}

response = requests.put(f"{BASE_URL}/products/20", json=updated_product)
print(response.json())
```

## PATCH

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

updated_product = {
    "category": 'electronic'
}

response = requests.patch(f"{BASE_URL}/products/19", json=updated_product)
print(response.json())
```

## DELETE

```python
import requests

BASE_URL = 'https://fakestoreapi.com'

response = requests.delete(f"{BASE_URL}/products/21")
print(response.json())
```
