Creamos de nuevo la tabla HC_CONSUMO añadiendo las claves foráneas

```python
%%sql
CREATE TABLE HC_CONSUMO (
    COD_CLIENTE INT NOT NULL,
    COD_TARIFA INT,
    COD_PERIODO INT,
    CONSUMO DECIMAL(8,2),
    PRIMARY KEY(COD_CLIENTE, COD_PERIODO),
    FOREIGN KEY(COD_CLIENTE) REFERENCES DM_CLIENTE (COD_CLIENTE),
    FOREIGN KEY(COD_TARIFA) REFERENCES DM_TARIFA (COD_TARIFA),
    FOREIGN KEY(COD_PERIODO) REFERENCES DM_PERIODO (COD_PERIODO)
);
```

Consultamos las claves foráneas

```python
%sql PRAGMA foreign_key_list('HC_CONSUMO')
```

```python
%sql PRAGMA table_info('HC_CONSUMO');
```

En SQLite, PRAGMA es una instrucción que se utiliza para interactuar con parámetros y configuraciones específicos de la base de datos. No es una operación SQL estándar, sino que es una extensión propia de SQLite que proporciona acceso a información y configuraciones adicionales. El propósito principal de PRAGMA es permitir a los usuarios y aplicaciones obtener o modificar ciertos aspectos de la configuración de la base de datos, ajustar el comportamiento del motor SQLite y obtener información sobre la base de datos y sus objetos

Creamos un campo calculado llamado grupo_tarifa que toma el valor ‘G1’ para las tarifas 100 y 102 y ‘G2’ para el resto:

```python
%%sql
select cod_cliente, cod_tarifa, consumo,
case when (cod_tarifa in (100,101)) then 'G1' else 'G2' end  grupo_tarifa
from HC_CONSUMO where cod_periodo = 201801;
```

Creamos un campo calculado llamado nivel_consumo  que toma los siguientes valores:  

Consumo < 10 -> 1  

Consumo >= 10 and consumo < 20 -> 2  

Consumo >=20 and consumo < 30 -> 3  

Resto -> 4


```python
%%sql
select cod_cliente, cod_tarifa, consumo,
case
when (consumo < 10) then 1
when (consumo >= 10 and consumo < 20) then 2
when (consumo >= 20 and consumo < 30) then 3
else 4
end  nivel_consumo
from HC_CONSUMO where cod_periodo = 201801;
```

```python
%%sql
WITH
    CONSUMO_CLIENTE as (select a.cod_cliente, a.cod_tarifa, a.cod_periodo, a.consumo,  b.nombre, b.tipo_cliente
    from HC_CONSUMO a inner join DM_CLIENTE b
    on a.cod_cliente = b.cod_cliente)
SELECT tipo_cliente, sum(consumo)
FROM CONSUMO_CLIENTE GROUP BY tipo_cliente
```
