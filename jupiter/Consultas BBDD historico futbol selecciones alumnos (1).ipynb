```python
%%sql
create table t_spain_matches as (
select g_date, tournament, home_team, away_team, home_score, away_score,
case
    when (home_team = 'Spain' and home_score > away_score) then 'victoria'
    when (away_team = 'Spain' and home_score < away_score) then 'victoria'
    when (home_team = 'Spain' and home_score < away_score) then 'derrota'
    when (away_team = 'Spain' and home_score > away_score) then 'derrota'
    when home_score = away_score then 'empate'
    else 'desconocido'
end as resultado,
case
    when (home_team = 'Spain') then home_score
    when (away_team = 'Spain') then away_score
    else 0
end  as goles_españa,
case
    when (home_team = 'Spain') then away_score
    when (away_team = 'Spain') then home_score
    else 0
end  as goles_contrario
FROM T_RESULTS
where (home_team = 'Spain' or away_team = 'Spain')
and g_date > '2000-01-01'
)
```

### Consultas OLAP

Tenemos varios tipos:   
* __ROLLUP__: Esta cláusula se utiliza para calcular subtotales y totales generales en agregaciones. Genera filas adicionales que contienen subtotales de los datos, con base en las columnas especificadas en la consulta. Ejemplo si pongo producto y región, primero me da el agregado por producto y región y luego por producto para todas las regiones. El orden de la jerarqía lo marca el orden en el que especifico las columnas en el group by
* __CUBE__: Similar a ROLLUP, pero genera todas las combinaciones posibles de subtotales. Ejemplo si pongo producto y región, primero me da el agregado por producto y región , luego por producto para todas las regiones y también daría por región para todos los productos. Si pongo tres campos en el group by, haría agregado por los 3 campos, todos los posibles agregados cogiendo campos de 2 en 2 y finalmente de 1 en 1.
* __GROUPING SETS___: Esta opción permite especificar conjuntos de columnas para los cuales se deben calcular los subtotales. Es más flexible que ROLLUP y CUBE, ya que te permite especificar combinaciones específicas de columnas para las cuales deseas los subtotales

```python
%%sql
select tournament, country, count(*) as num_partidos,
GROUPING(tournament) AS grp_tournament,
GROUPING(country) AS grp_country
FROM T_RESULTS
where country in ('Spain', 'France', 'Germany', 'Italy')
and ((tournament like '%FIFA%') or (tournament like '%UEFA%'))
group by tournament, country
with ROLLUP
```

```python
%%sql
select tournament, country, count(*) as num_partidos
FROM T_RESULTS
where country in ('Spain', 'France', 'Germany', 'Italy')
and ((tournament like '%FIFA%') or (tournament like '%UEFA%'))
group by tournament, country
with CUBE(tournament, country)
```

```python
%%sql
select g_date, home_team, away_team, home_score, away_score, tournament,
FIRST_VALUE(away_team) OVER  w  as first_team,
FIRST_VALUE(away_team) over wa as last_team
from T_RESULTS
where g_date >= '2023-01-01' and tournament like '%FIFA%'
window w as (   PARTITION BY home_team  ORDER BY g_date ),
 wa as (   PARTITION BY home_team  ORDER BY g_date desc)
```

Para hacerlo con last_value hay que hacer un pequeño cambio: The default windowing clause is "RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW". Altering the windowing clause to "ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING" gives us the result we probably expected.

Cuando incluímos la ventana sin ordenación (   PARTITION BY home_team ) los cálculos (sum, avg, last_value...) los hace
para la ventana completa, si incluísmo la ordenación (   PARTITION BY home_team  ORDER BY g_date ) los  hace desde la
primera fila de la ventana hasta la actual (CURRENT ROW)

The bounds can be any of these five options:

* UNBOUNDED PRECEDING – All rows before the current row.
* n PRECEDING – n rows before the current row.
* CURRENT ROW – Just the current row.
* n FOLLOWING – n rows after the current row.
* UNBOUNDED FOLLOWING – All rows after the current row.

```python
%%sql
select g_date, home_team, away_team, home_score, away_score, tournament,
first_value(away_team) OVER (   PARTITION BY home_team  ORDER BY g_date ) as first_team,
last_value(away_team) OVER (   PARTITION BY home_team  ORDER BY g_date rows BETWEEN
           UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING ) as last_team
from T_RESULTS
where (g_date >= '2021-01-01' and g_date <= '2021-12-31')
order by home_team, g_date
limit 20
```

```python
%%sql
with cte as
(select *
from T_RESULTS
where (home_team, away_team) in 'Spain')
select * from cte where home_score >= 4
```

```python
%%sql
with cte as
(
select  home_team,
first_value(away_team) OVER (   PARTITION BY home_team  ORDER BY g_date ) as first_team,
first_value(away_team) OVER (   PARTITION BY home_team  ORDER BY g_date desc ) as last_team
from T_RESULTS
where (g_date >= '2021-01-01' and g_date <= '2021-12-31')
order by home_team, g_date
)
select home_team, first_team, last_team
from cte
group by home_team
limit 10
```

Para hacerla en MySQL usamos CTE recursivos

Uso de CTE recursivo. En la construcción de la tabla se va llamando a si misma

```python
%%sql
WITH RECURSIVE cte AS
(
  -- inicio
  SELECT 1 AS n
  UNION ALL
  -- recursividad
  SELECT n + 1 FROM cte WHERE n < 5
)
SELECT * FROM cte
```

```python
%%sql
WITH RECURSIVE jerarquia(id_empleado, nombre_empleado, responsable, nombre_responsable, jefe, nombre_jefe)
AS(
    -- Valores iniciales. Corresponde a la selección del responsable
    SELECT empleados.id_empleado, empleados.nombre_empleado,
        0 AS responsable, cast('' as char(100)) AS nombre_responsable,
        empleados.id_empleado AS jefe, empleados.nombre_empleado AS nombre_jefe
    FROM  empleados
    WHERE id_empleado = 1
    UNION
    -- Recursividad: Subconsulta recursiva para todos los empleados bajo la jerarquia del responsable
    SELECT empleados.id_empleado, empleados.nombre_empleado,
        jerarquia.id_empleado AS responsable, jerarquia.nombre_empleado AS nombre_responsable,
        jerarquia.jefe, jerarquia.nombre_jefe
    FROM empleados, jerarquia
    WHERE empleados.responsable = jerarquia.id_empleado
)
select id_empleado, nombre_empleado, responsable, nombre_responsable, jefe, nombre_jefe FROM jerarquia
```

Los puntos claves de la consulta son:
* Consulta de definición de los valores iniciales dentro de la CTE : Sirve para determinar los valores iniciales de la recursión , en este caso, los del responsable de primer nivel con ID_EMPLEADO = 1 (línea 9)
* Dentro de la parte de la consulta con recursividad, entra por un valor de id_empleado y lo busca como responsable (WHERE empleados.responsable = jerarquia.id_empleado, por ejemplo id_empleado = 1). En los datos obtenidos en la SELECT  empleados.id_empleado ((que sería id_empleado = 4) es el empleado que está a cargo del jerarquia.id_empleado de entrada  y jerarquia.empleado es el responsable (1).   Los resultados de la SELECT son mapeados a los de la definición de la CTE (primera línea jerarquia(id_empleado, nombre_empleado,..), de forma que el id_empleado retornado (4)  pasa a ser el id_empleado de jerarquia por el que hace la siguiente búsqueda que va a encontrar 6 y así sucesivamente hasta que se llegue a un id_empleado que no es a su vez responsable

Un ejemplo de tratamiento de query recursiva es que se se realiza para aplanar la __jerarquía padre-hijo contable de SAP__ para con una __estructura aplanada pasarla a BigQuery__
