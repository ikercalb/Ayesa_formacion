## EJERCICIO 2

- Genera una matriz de tamaño 6x6 con valores aleatorios enteros entre 1 y 100.
- Filtra los valores mayores que 50 y reemplázalos por 0.
- Calcula la suma de los valores en cada fila y columna.


```python
matriz_aleatoria = np.random.rand(5, 5)

media = np.mean(matriz_aleatoria)
desviacion_estandar = np.std(matriz_aleatoria)
mediana = np.median(matriz_aleatoria)

valor_maximo = np.max(matriz_aleatoria)
valor_minimo = np.min(matriz_aleatoria)

suma_total = np.sum(matriz_aleatoria)

# Mostrar los resultados
print("Matriz Aleatoria de 5x5:")
print(matriz_aleatoria)
print("\nMedia de la Matriz:", media)
print("Desviación Estándar de la Matriz:", desviacion_estandar)
print("Mediana de la Matriz:", mediana)
print("Valor Máximo en la Matriz:", valor_maximo)
print("Valor Mínimo en la Matriz:", valor_minimo)
print("Suma Total de la Matriz:", suma_total)
```

```python
matriz_enteros = np.random.randint(1, 101, size=(6, 6))

matriz_enteros[matriz_enteros > 50] = 0

suma_filas = np.sum(matriz_enteros, axis=1)
suma_columnas = np.sum(matriz_enteros, axis=0)

# Mostrar los resultados
print("Matriz de 6x6 con Valores Aleatorios Enteros entre 1 y 100:")
print(matriz_enteros)
print("\nSuma de Valores en Cada Fila:")
print(suma_filas)
print("\nSuma de Valores en Cada Columna:")
print(suma_columnas)
```

```python
def notas_aprobados(notas):
    serie_notas = pd.Series(notas)
    aprobados = serie_notas[serie_notas >= 6]
    aprobados_ordenados = aprobados.sort_values(ascending=False)

    return aprobados_ordenados

notas_alumnos = {
    'Adrian': 8.5,
    'Pablo': 5.0,
    'Manolito': 7.0,
    'Antoñito': 4.5,
    'Manuel': 9.0,
    'Javi': 6.5
}

notas_aprobadas = notas_aprobados(notas_alumnos)
print(notas_aprobadas)
```

```python
meses = ["Enero", "Febrero"]

def balance (meses, contabilidad):
    contabilidad["Balance"] = contabilidad["Ventas"] - contabilidad["Gastos"]
    filtro = contabilidad[contabilidad["Mes"].isin(meses)]
    return filtro.Balance.sum()

print(balance(meses, contabilidad))
```

```python
def resumen_cotizaciones(fichero):
    df = pd.read_csv(fichero, sep = ";", decimal = ",", thousands='.', index_col=0)
    return pd.DataFrame([df.min(), df.max(), df.mean()], index=['Mínimo', 'Máximo', 'Media'])

resumen_cotizaciones('cotizacion.csv')
```
